**feature scaling**

```python
list_1=[]
for i in churn_df.columns:
    list_1.append(i)
from sklearn.preprocessing import MinMaxScaler
for i in list_1:
    MMS=MinMaxScaler()
    churn_df[i]=MMS.fit_transform(churn_df[[i]])
```

```python
churn_df
```

**inorder to get faster convergence to the global minima and reduce the training time we will implement feature scaling**

```python
X=churn_df.drop("Exited",axis=1)
y=churn_df["Exited"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1234)
```

**creating a neural network**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LeakyReLU,ReLU,Dropout

```

```python
model=Sequential()
```

```python
model.add(Dense(units=10,activation='relu'))   # input layer
```

```python
model.add(Dense(units=5,activation='relu'))  # hidden layer 1
```

```python
model.add(Dense(units=4,activation='relu'))   # hideen layer 2
```

```python
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# adding your learning rate
# tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
history=model.fit(X_train,y_train,validation_split=0.33,batch_size=15,epochs=50)
```

```python
model.summary()
```

```python
history.history.keys()
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("model accuracy")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model accuracy")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
y_pred=model.predict(X_test)
y_pred=y_pred>0.5
y_pred=np.where(y_pred,1,0)
```

**evaluation**

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
ac_s=accuracy_score(y_test,y_pred)
print("The accuracy",ac_s)
ps=precision_score(y_test,y_pred)
print("The precision",ps)
rs=recall_score(y_test,y_pred)
print("recall score",rs)
f1=f1_score(y_test,y_pred)
print("f1 score",f1)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,roc_auc_score
cm=confusion_matrix(y_test,y_pred)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,
                            display_labels = [False, True])
disp.plot()
plt.grid(False)
plt.show()


tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
print("True negative:",tn)
print("False postive:",fp)
print("False negative:",fn)
print("True postive:",tp)
```
